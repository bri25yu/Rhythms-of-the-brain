```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from math import sin, pi
import random
```

```python

def f(x,t): # define the function f(x,t)
    return np.cos(t)

epsilon = 0.1 # define timestep
t = np.arange(0,10,epsilon) # define an array for t
x = np.zeros(t.shape) # define an array for x
x[0]= 0 # set initial condition
for i in range(1,t.shape[0]):
    x[i] = epsilon*f(x[i-1],t[i-1])+x[i-1] # Euler Integration Step
```

```python
x
```

```python
plt.plot(x)
```

```python
plt.plot(x-np.sin(t))
```

#Runge_Kutta Second Order

```python
def f(y,t): # define the function f(x,t)
    return np.cos(t)

epsilon = 0.1 # define timestep
t = np.arange(0,10,epsilon) # define an array for t
y = np.zeros(t.shape) # define an array for x
y[0]= 1 # set initial condition
for i in range(1,t.shape[0]):
    k1=(1/2)*epsilon*f(y[i-1],t[i-1])
    y[i] = y[i-1]+epsilon*f(y[i-1]+k1,t[i-1]+epsilon/2) # Runge-Kutta 2nd Order Integration Step
```

```python
y
```

```python
plt.plot(y)
```

```python
plt.plot(y-np.sin(t))
```
