## Euler's Method:

```python

def f(x,t): # define the function f(x,t)
    return np.exp(t)

epsilon = 0.001 # define timestep
ex_t = np.arange(0,5, epsilon) # define an array for t
ex_x = np.zeros(ex_t.shape) # Initializing an array for x
ex_x[0]= 1 # set initial condition
for i in range(1,ex_t.shape[0]):
    ex_x[i] = epsilon*f(ex_x[i-1],ex_t[i-1])+ex_x[i-1] # Euler Integration Step
```

## Runge-Kutta Second Order:

```python
def f(y,t): # define the function f(x,t)
    return 5*y

epsilon = 0.001 # define timestep
ex_r_t = np.arange(0,2,epsilon) # define an array for t
ex_y = np.zeros(ex_r_t.shape) # define an array for x
ex_y[0]= 1 # set initial condition
for i in range(1,ex_r_t.shape[0]):
    k1=(1/2)*epsilon*f(ex_y[i-1],ex_r_t[i-1])
    ex_y[i] = ex_y[i-1]+epsilon*f(ex_y[i-1]+k1,ex_r_t[i-1]) # Runge-Kutta 2nd Order Integration Step
```
